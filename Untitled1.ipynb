```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('https://raw.githubusercontent.com/wiilldii/study/ab33f5f8bb1f5b874c77d139a73efde320236a2d/stock_list.csv', sep=',')

sns.set_theme(style="whitegrid")

#  1 ------------------------------
median_value = data['MarketCapitalization'].median()

bins = [0, 1e9, 1e12, 1e15]  #    
labels = ['Small Cap (<1B)', 'Mid Cap (1B-1T)', 'Large Cap (>1T)']
data['MarketCapCategory'] = pd.cut(data['MarketCapitalization'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(12, 8))
sns.histplot(data['MarketCapitalization'], bins=50, kde=True, color='blue', log_scale=(True, False),  alpha=0.7)
plt.axvline(median_value, color='green', linestyle='--', label=f" : {median_value:,.0f}")

plt.title("  ", fontsize=16, weight='bold')
plt.xlabel("   ", fontsize=12)
plt.ylabel(" ", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#  2 -----------------------------

plt.figure(figsize=(14, 8))

sns.scatterplot(x=data['Close'], y=data['MarketCapitalization'], size=data['IssuedShares'], alpha=0.6, sizes=(20, 200), edgecolor='w', hue=data['Section/Products'], palette='viridis')

top_5_companies = data.nlargest(5, 'MarketCapitalization')
for _, row in top_5_companies.iterrows():
    plt.text(row['Close'], row['MarketCapitalization'], row['Name'], fontsize=5, weight='bold', color='black')

plt.xscale('log')
plt.yscale('log')

plt.title("  vs  ", fontsize=16, weight='bold')
plt.xlabel("  ( )", fontsize=12)
plt.ylabel("  ( )", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout


#    ------------------------

median_issued = data['IssuedShares'].median()
std_dev_issued = data['IssuedShares'].std()

bins = [0, 1e7, 1e9, 1e11]
labels = [' (<10)', ' (10-1B)', ' (>1B)']
data['IssuedSharesCategory'] = pd.cut(data['IssuedShares'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(12, 8))
hist_plot = sns.histplot(data=data, x='IssuedShares', hue='IssuedSharesCategory', bins=50, kde=True, palette='Set2', log_scale=(True, False), alpha=0.7)

plt.axvline(median_issued, color='blue', linestyle='--', label=f": {median_issued:,.0f}")

plt.title("   ", fontsize=16, weight='bold')
plt.xlabel("   ( )", fontsize=12)
plt.ylabel("", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()




#----------------------------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['IssuedShares'], y=data['MarketCapitalization'], alpha=0.6, color='purple', edgecolor='w')
plt.title("Issued Shares vs Market Capitalization", fontsize=16, weight='bold')
plt.xlabel("Issued Shares", fontsize=12)
plt.ylabel("Market Capitalization", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()

filtered_data = data[data['33SectorName'] != '-']

sector_volatility = filtered_data.groupby('33SectorName')['Close'].std().sort_values()

top_3 = sector_volatility.head(3)
middle_index = len(sector_volatility) // 2
middle_2 = sector_volatility.iloc[middle_index - 1:middle_index + 1]
bottom_3 = sector_volatility.tail(3)

selected_sectors = pd.concat([top_3, middle_2, bottom_3]).reset_index()
selected_sectors.columns = ['Sector', 'Volatility']

plt.figure(figsize=(10, 6))
sns.barplot(x='Volatility', y='Sector', data=selected_sectors, palette='coolwarm')
plt.title("Selected Sectors by Volatility (Top, Middle, Bottom)")
plt.xlabel("Volatility (Standard Deviation of Close Price)")
plt.ylabel("Sector")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```
